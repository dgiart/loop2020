```python
#Re=55, vac-loop,eps=1  OLD
import numpy as np
import matplotlib.pyplot as plt
from time import time
%matplotlib inline
l=np.linspace(1.,3.5,15)
R=10**l
R=np.append([6,7,8,9],R)
print(f'R: {R}')
Zv=[]
Zi=[]
start=time()
for r in R:
    t=time()
    zv=LiebmanAniz('exp',1.,5,55,r, P=12., _w=1.95, dr=.5,dz=.5,form='iz')
    Zv.append(zv['flow'])
    zi=LiebmanAniz('exp',1.,5,55,r, P=-24, _w=1.95, dr=.5,dz=.5,form='iz')
    Zi.append(zi['flow'])
    print (f"R={r}, Zv={zv['flow']}, Zi={zi['flow']},B={1-zv['flow']/zi['flow']}, T={time()-t}")
    print('\n *****************\n')     
           
Zv55v12=np.array(Zv)
Zi55v12=np.array(Zi)
B55v12=1-Zv55v12/Zi55v12
plt.plot(R,B55v12,'*')
plt.semilogx()
plt.show()
pickle.dump( Zv55v12, open( "RESULTS/Zv55v12.txt", "wb" ) )
pickle.dump( Zi55v12, open( "RESULTS/Zi55v12.txt", "wb" ) )
pickle.dump( B55v12, open( "RESULTS/B55v12.txt", "wb" ) )
print (f'DONE {time()-start}')
    
```

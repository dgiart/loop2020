```python
import numpy as np
import matplotlib.pyplot as plt
from time import time
%matplotlib inline
l=np.linspace(1.,3.5,15)
R=10**l
R=np.append([6,7,8,9],R)
print(f'R: {R}')
Zv=[]
Zi=[]
start=time()
for r in R:
    t=time()
    zv=LiebmanAniz('one',0.8,5,55,r, P=-179.073, _w=1.95, dr=.5,dz=.5,form='aniz')
    Zv.append(zv['flow'])
    zi=LiebmanAniz('one',0.8,5,55,r, P=358.146, _w=1.95, dr=.5,dz=.5,form='aniz')
    Zi.append(zi['flow'])
    print (f"R={r}, Zv={zv['flow']}, Zi={zi['flow']},B={1-zv['flow']/zi['flow']}, T={time()-t}")
           
           
Zv=np.array(Zv)
Zi=np.array(Zi)
B=1-Zv/Zi
plt.plot(R,B,'*')
plt.semilogx()
plt.show()
print (f'DONE {time()-start}')
    
```

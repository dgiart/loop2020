```python
import sys
import pickle
import numpy as np
sys.path.append('../')
from l2020_iz_anizotropic import LiebmanAniz
```

```python
#B from Re,R=3000 vac-loop,eps=1.0
import numpy as np
import matplotlib.pyplot as plt
from time import time
%matplotlib inline
R=3000
Re=np.arange(92,128,3)
print(f'R: {Re}')
Zv=[]
Zi=[]
start=time()
for re in Re:
    t=time()
    zv=LiebmanAniz('one',1.,5,re,R, P=-179.073, _w=1.95, dr=.5,dz=.5,form='aniz')
    Zv.append(zv['flow'])
    zi=LiebmanAniz('one',1.,5,re,R, P=358.146, _w=1.95, dr=.5,dz=.5,form='aniz')
    Zi.append(zi['flow'])
    print (f"Re={re}, Zv={zv['flow']}, Zi={zi['flow']},B={1-zv['flow']/zi['flow']}, T={time()-t}")
    print('\n *****************\n')     
           
ZvR3000v10=np.array(Zv)
ZiR3000v10=np.array(Zi)
BR3000v10=1-ZvR3000v10/ZiR3000v10
plt.plot(Re,BR3000v10,'*')
# plt.semilogx()
plt.show()
pickle.dump( ZvR3000v10, open( "RESULTS/ZvR3000v10.txt", "wb" ) )
pickle.dump( ZiR3000v10, open( "RESULTS/ZiR3000v10.txt", "wb" ) )
pickle.dump( BR3000v10, open( "RESULTS/BR3000v10.txt", "wb" ) )
print (f'DONE {time()-start}')
    
```

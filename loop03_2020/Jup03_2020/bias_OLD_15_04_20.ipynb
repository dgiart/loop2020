```python
import sys
import pickle
import numpy as np
sys.path.append('../')
from l2020_iz_anizotropic import LiebmanAniz
```

```python
import numpy as np
import matplotlib.pyplot as plt
from time import time
%matplotlib inline
l=np.linspace(1.,3.5,15)
R=10**l
R=np.append([6,7,8,9],R)
R=[int(r) for r in R]
print(f'R: {R}')
Zv=[]
Zi=[]
start=time()
for r in R:
    t=time()
    zv=LiebmanAniz('exp',1.,3,55,r, P=-12., _w=1.95, dr=.5,dz=.5,form='iz')
    Zv.append(zv['flow'])
    zi=LiebmanAniz('exp',1.,3,55,r, P=24., _w=1.95, dr=.5,dz=.5,form='iz')
    Zi.append(zi['flow'])
    print (f"R={r}, Zv={zv['flow']}, Zi={zi['flow']},B={1-zv['flow']/zi['flow']}, T={time()-t}")
    print('\n *****************\n')     
           
Zv55=np.array(Zv)
Zi55=np.array(Zi)
B55=1-Zv55/Zi55
plt.plot(R,B55,'*')
plt.semilogx()
plt.show()
pickle.dump( Zv55, open( "Zv55vturkin.txt", "wb" ) )
pickle.dump( Zi55, open( "Zi55vturkin.txt", "wb" ) )
pickle.dump( B55, open( "B55vturkin.txt", "wb" ) )
print (f'DONE {time()-start}')
```

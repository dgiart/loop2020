```python
import sys
import pickle
import numpy as np
sys.path.append('../')
from l2020_iz_anizotropic import LiebmanAniz
```

```python
import numpy as np
import matplotlib.pyplot as plt
from time import time
%matplotlib inline
l=np.linspace(1.,3.5,15)
R=10**l
R=np.append([6,7,8,9],R)
R=[int(r) for r in R]
print(f'R: {R}')
Zv=[]
Zi=[]
start=time()
for r in R:
    t=time()
    zv=LiebmanAniz('exp',1.,3,300,r, P=-12., _w=1.99, dr=.5,dz=.5,form='iz')
    Zv.append(zv['flow'])
    zi=LiebmanAniz('exp',1.,3,300,r, P=24., _w=1.99, dr=.5,dz=.5,form='iz')
    Zi.append(zi['flow'])
    print (f"R={r}, Zv={zv['flow']}, Zi={zi['flow']},B={1-zv['flow']/zi['flow']}, T={time()-t}")
    print('\n *****************\n')     
           
Zv300=np.array(Zv)
Zi300=np.array(Zi)
B300=1-Zv300/Zi300
plt.plot(R,B300,'*')
plt.semilogx()
plt.show()
pickle.dump( Zv300, open( "Zv300turkin.txt", "wb" ) )
pickle.dump( Zi300, open( "Zi300turkin.txt", "wb" ) )
pickle.dump( B300, open( "B300turkin.txt", "wb" ) )
print (f'DONE {time()-start}')
```

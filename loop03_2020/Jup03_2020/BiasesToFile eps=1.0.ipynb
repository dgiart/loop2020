```python
import sys
import pickle
import numpy as np
sys.path.append('../')
from l2020_iz_anizotropic import LiebmanAniz
```

```python
#Re=55, vac-loop,eps=1.0
import numpy as np
import matplotlib.pyplot as plt
from time import time
%matplotlib inline
l=np.linspace(1.,3.5,15)
R=10**l
R=np.append([6,7,8,9],R)
print(f'R: {R}')
Zv=[]
Zi=[]
start=time()
for r in R:
    t=time()
    zv=LiebmanAniz('one',1.,5,55,r, P=-179.073, _w=1.95, dr=.5,dz=.5,form='aniz')
    Zv.append(zv['flow'])
    zi=LiebmanAniz('one',1.,5,55,r, P=358.146, _w=1.95, dr=.5,dz=.5,form='aniz')
    Zi.append(zi['flow'])
    print (f"R={r}, Zv={zv['flow']}, Zi={zi['flow']},B={1-zv['flow']/zi['flow']}, T={time()-t}")
    print('\n *****************\n')     
           
Zv55v10=np.array(Zv)
Zi55v10=np.array(Zi)
B55v10=1-Zv55v10/Zi55v10
plt.plot(R,B55v10,'*')
plt.semilogx()
plt.show()
pickle.dump( Zv55v10, open( "RESULTS/Zv55v10.txt", "wb" ) )
pickle.dump( Zi55v10, open( "RESULTS/Zi55v10.txt", "wb" ) )
pickle.dump( B55v10, open( "RESULTS/B55v10.txt", "wb" ) )
print (f'DONE {time()-start}')
    
```

```python
import numpy as np
import matplotlib.pyplot as plt
from time import time
%matplotlib inline
l=np.linspace(1.,3.5,15)
R=10**l
R=np.append([6,7,8,9],R)
print(f'R: {R}')
Zv=[]
Zi=[]
start=time()
for r in R:
    t=time()
    zv=LiebmanAniz('exp',3,55,int(r), P=-12., _w=1.95, dr=1.,dz=1.,form='iz')
    Zv.append(zv['flow'])
    zi=LiebmanAniz('exp',3,55,int(r), P=24., _w=1.95, dr=0.5,dz=0.5,form='iz')
    Zi.append(zi['flow'])
    print (f"R={r}, Zv={zv['flow']}, Zi={zi['flow']}, T={time()-t}")
           
           
Zv=np.array(Zv)
Zi=np.array(Zi)
B=1-Zv/Zi
plt.plot(R,B,'*')
plt.semilogx()
plt.show()
print (f'DONE {time()-start}')
    
```
